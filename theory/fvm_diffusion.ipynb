# A.1 Two-Dimensional Neutron Diffusion Calculation

## Governing Equation

We consider the two-dimensional, one-group neutron diffusion equation
without anisotropic scattering:

$$
\frac{\partial}{\partial x}
\left(
- D(x,y)\frac{\partial \phi(x,y)}{\partial x}
\right)
+
\frac{\partial}{\partial y}
\left(
- D(x,y)\frac{\partial \phi(x,y)}{\partial y}
\right)
+
\Sigma_t(x,y)\phi(x,y)
=
\frac{\nu\Sigma_f(x,y)}{k_{\mathrm{eff}}}\phi(x,y)
+
\Sigma_s(x,y)\phi(x,y)
\tag{1}
$$

where:

- $\phi(x,y)$ : scalar neutron flux  
- $D(x,y)$ : diffusion coefficient  
- $\Sigma_t$ : total cross section  
- $\Sigma_s$ : scattering cross section  
- $\nu\Sigma_f$ : fission production term  
- $k_{\mathrm{eff}}$ : effective multiplication factor  

---

## Spatial Discretization

The computational domain is divided into
$N_x$ cells in the $x$ direction and
$N_y$ cells in the $y$ direction.

Let $(i,j)$ denote the mesh cell centered at
$(x_i, y_j)$ with widths $\Delta x_{i,j}$ and $\Delta y_{i,j}$.

Integrating Equation (1) over cell $(i,j)$ yields
the neutron balance equation:

$$
\frac{\bar{J}^{x+}_{i,j}-\bar{J}^{x-}_{i,j}}{\Delta x_{i,j}}
+
\frac{\bar{J}^{y+}_{i,j}-\bar{J}^{y-}_{i,j}}{\Delta y_{i,j}}
+
\Sigma_{t,i,j}\bar{\phi}_{i,j}
=
\frac{\nu\Sigma_{f,i,j}}{k_{\mathrm{eff}}}\bar{\phi}_{i,j}
+
\Sigma_{s,i,j}\bar{\phi}_{i,j}
\tag{2}
$$

where:

- $\bar{\phi}_{i,j}$ : cell-averaged flux  
- $\bar{J}^{x\pm}_{i,j}$ : averaged neutron current in $x$ direction  
- $\bar{J}^{y\pm}_{i,j}$ : averaged neutron current in $y$ direction  

---

## Interface Flux Approximation

Consider the interface between cells $(i-1,j)$ and $(i,j)$.

The neutron current is approximated by:

$$
\bar{J}^{x-}_{i,j}
\approx
- D_{i-1,j}
\frac{\phi_s-\bar{\phi}_{i-1,j}}{\Delta x_{i-1,j}/2}
=
- D_{i,j}
\frac{\bar{\phi}_{i,j}-\phi_s}{\Delta x_{i,j}/2}
\tag{3}
$$

where $\phi_s$ is the interface flux.

---

## Interface Flux Solution

From Equation (3), the interface flux is given by:

$$
\phi_s
=
\frac{
D_{i-1,j}\bar{\phi}_{i-1,j}\Delta x_{i,j}
+
D_{i,j}\bar{\phi}_{i,j}\Delta x_{i-1,j}
}{
D_{i-1,j}\Delta x_{i,j}
+
D_{i,j}\Delta x_{i-1,j}
}
\tag{4}
$$

---

## Neutron Current Formulation

Substituting Equation (4) into Equation (3) yields:

$$
\bar{J}^{x-}_{i,j}
=
-
\frac{
2D_{i-1,j}D_{i,j}
}{
D_{i-1,j}\Delta x_{i,j}
+
D_{i,j}\Delta x_{i-1,j}
}
(\bar{\phi}_{i,j}-\bar{\phi}_{i-1,j})
\tag{5}
$$

Similarly, other currents are approximated as:

$$
\bar{J}^{x+}_{i,j}
=
-
\frac{
2D_{i,j}D_{i+1,j}
}{
D_{i,j}\Delta x_{i+1,j}
+
D_{i+1,j}\Delta x_{i,j}
}
(\bar{\phi}_{i+1,j}-\bar{\phi}_{i,j})
\tag{6}
$$

$$
\bar{J}^{y-}_{i,j}
=
-
\frac{
2D_{i,j-1}D_{i,j}
}{
D_{i,j-1}\Delta y_{i,j}
+
D_{i,j}\Delta y_{i,j-1}
}
(\bar{\phi}_{i,j}-\bar{\phi}_{i,j-1})
\tag{7}
$$

$$
\bar{J}^{y+}_{i,j}
=
-
\frac{
2D_{i,j}D_{i,j+1}
}{
D_{i,j}\Delta y_{i,j+1}
+
D_{i,j+1}\Delta y_{i,j}
}
(\bar{\phi}_{i,j+1}-\bar{\phi}_{i,j})
\tag{8}
$$

---

## Five-Point Difference Scheme

Substituting Equations (5)–(8) into Equation (2),
we obtain the five-point finite difference scheme:

$$
A^{x-}_{i,j}\bar{\phi}_{i-1,j}
+
A^{y-}_{i,j}\bar{\phi}_{i,j-1}
+
A^{0}_{i,j}\bar{\phi}_{i,j}
+
A^{x+}_{i,j}\bar{\phi}_{i+1,j}
+
A^{y+}_{i,j}\bar{\phi}_{i,j+1}
=
\frac{\nu\Sigma_{f,i,j}}{k_{\mathrm{eff}}}\bar{\phi}_{i,j}
\tag{9}
$$

---

## Coefficient Definitions

The coefficients in Equation (9) are defined as:

$$
A^{x-}_{i,j}
=
-
\frac{
2D_{i-1,j}D_{i,j}
}{
\left(
D_{i-1,j}\Delta x_{i,j}
+
D_{i,j}\Delta x_{i-1,j}
\right)\Delta x_{i,j}
}
\tag{10}
$$

$$
A^{y-}_{i,j}
=
-
\frac{
2D_{i,j-1}D_{i,j}
}{
\left(
D_{i,j-1}\Delta y_{i,j}
+
D_{i,j}\Delta y_{i,j-1}
\right)\Delta y_{i,j}
}
\tag{11}
$$

$$
A^{x+}_{i,j}
=
-
\frac{
2D_{i,j}D_{i+1,j}
}{
\left(
D_{i,j}\Delta x_{i+1,j}
+
D_{i+1,j}\Delta x_{i,j}
\right)\Delta x_{i,j}
}
\tag{12}
$$

$$
A^{y+}_{i,j}
=
-
\frac{
2D_{i,j}D_{i,j+1}
}{
\left(
D_{i,j}\Delta y_{i,j+1}
+
D_{i,j+1}\Delta y_{i,j}
\right)\Delta y_{i,j}
}
\tag{13}
$$

$$
A^{0}_{i,j}
=
\Sigma_{a,i,j}
-
\left(
A^{x-}_{i,j}
+
A^{y-}_{i,j}
+
A^{x+}_{i,j}
+
A^{y+}_{i,j}
\right)
\tag{14}
$$

where

$$
\Sigma_{a,i,j}=\Sigma_{t,i,j}-\Sigma_{s,i,j}.
$$

---

## Summary

The above formulation leads to a standard five-point
finite difference discretization of the
two-dimensional neutron diffusion equation.

This scheme conserves neutron balance in each cell
and forms the basis for eigenvalue calculations
of $k_{\mathrm{eff}}$.
