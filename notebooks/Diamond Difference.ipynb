# 1D Weighted Diamond Difference (WDD) Method

## Overview

The Weighted Diamond Difference (WDD) method is a spatial discretization
scheme for the one-dimensional neutron transport equation.
It is based on enforcing neutron balance within each spatial mesh
while approximating the angular flux by a weighted average.

This method is an extension of the standard Diamond Difference (DD) scheme
and improves numerical stability by introducing a weighting parameter.

---

## Governing Transport Equation

Consider the steady-state one-dimensional transport equation
for discrete ordinate direction $m$:

$$
\mu_m \frac{d \psi_m(x)}{dx}
+ \Sigma_t(x)\psi_m(x)
= Q_m(x)
$$

where:

- $\psi_m(x)$ : angular flux for direction $m$
- $\mu_m$ : direction cosine
- $\Sigma_t$ : total cross section
- $Q_m$ : source term

---

## Mesh Integration

Integrating the transport equation over mesh cell $i$
with width $\Delta x_i$ gives:

$$
\mu_m(\psi_{m,i}^{in} - \psi_{m,i}^{out})
+ \Sigma_{t,i}\bar{\psi}_{m,i}\Delta x_i
=
\bar{Q}_{m,i}\Delta x_i
\tag{1}
$$

where:

- $\psi_{m,i}^{in}$ : incoming angular flux
- $\psi_{m,i}^{out}$ : outgoing angular flux
- $\bar{\psi}_{m,i}$ : cell-averaged angular flux
- $\bar{Q}_{m,i}$ : cell-averaged source
- $\Delta x_i$ : mesh width

Equation (1) represents the neutron balance within a mesh cell.

---

## Closure Relation

Equation (1) contains two unknowns,
$\psi_{m,i}^{out}$ and $\bar{\psi}_{m,i}$,
and therefore is not closed.

To close the system, the WDD method assumes:

$$
\bar{\psi}_{m,i}
=
(1-a)\psi_{m,i}^{in}
+
a\psi_{m,i}^{out},
\quad
\frac{1}{2} \le a < 1
\tag{2}
$$

where:

- $a$ : weighting parameter

This relation approximates the cell-average flux
by a weighted average of incoming and outgoing fluxes.

---

## Special Cases

### Diamond Difference (DD) Scheme

When

$$
a = \frac{1}{2},
$$

Equation (2) becomes:

$$
\bar{\psi}_{m,i}
=
\frac{1}{2}
(\psi_{m,i}^{in} + \psi_{m,i}^{out})
$$

This corresponds to the standard Diamond Difference method.

---

### Weighted Diamond Difference (WDD) Scheme

For

$$
a \ne \frac{1}{2},
$$

the scheme is called the Weighted Diamond Difference method.

Larger values of $a$ increase numerical stability
and help suppress negative fluxes,
especially in optically thick cells.

---

## Discrete Sweep Equation

Substituting Equation (2) into Equation (1) yields:

$$
\psi_{m,i}^{out}
=
\frac{
\mu_m + (1-a)\Sigma_{t,i}\Delta x_i
}{
\mu_m - a\Sigma_{t,i}\Delta x_i
}
\psi_{m,i}^{in}
+
\frac{
\Delta x_i
}{
\mu_m - a\Sigma_{t,i}\Delta x_i
}
\bar{Q}_{m,i}
\tag{3}
$$

This equation is used in the transport sweep
to compute the outgoing flux from the incoming flux.

---

## Physical Interpretation

The WDD method enforces neutron balance
within each spatial cell.

Key characteristics:

- Conservation of neutron balance
- Simple implementation
- Suitable for discrete ordinates methods
- Improved stability compared with DD

The weighting parameter $a$
controls the amount of upwinding in the scheme.

---

## Advantages and Limitations

### Advantages

- Conserves neutrons in each mesh
- Easy to implement
- More stable than DD
- Reduces negative fluxes

### Limitations

- Numerical diffusion for large $a$
- Reduced accuracy in optically thin regions
- Not strictly positivity-preserving

---

## Summary

The Weighted Diamond Difference method
is a conservative discretization scheme
for the one-dimensional transport equation.

By introducing a weighting parameter,
it extends the Diamond Difference method
and improves robustness in practical calculations.

It is widely used in discrete ordinates neutron transport solvers.
